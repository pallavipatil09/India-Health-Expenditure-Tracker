```python
import pandas as pd
import numpy as np
```

```python
health = pd.read_csv("India_health.csv")
imr = pd.read_excel("Infant_mortality_rate.xls",header=[2,3])
beds = pd.read_csv("Hospitals_and_beds.csv")
gsdp = pd.read_csv("gsdp_data.csv")
coords = pd.read_csv("state_coords.csv")
```

```python
print("Health:",health.shape)
print("IMR:",imr.shape)
print("Beds:",beds.shape)
print("GSDP:",gsdp.shape)
print("Coords:",coords.shape)
```

```python
health_clean = health.rename(columns={
    'State/UT': 'state',
    'Proportion of men and women Reporting Asthma 15-49 years - Men': 'asthma_men',
    'Proportion of men and women Reporting Asthma 15-49 years - Women': 'asthma_women',
    'Percentage of Grade-II disability cases amongst new detected cases of Leprosy - 2015': 'leprosy_2015',
    'Percentage of Grade-II disability cases amongst new detected cases of Leprosy - 2020 (Provisional)': 'leprosy_2020',
    'Percentage of population (15-24 years) with comprehensive knowledge of HIV/AIDS - Men': 'hiv_knowledge_men',
    'Percentage of population (15-24 years) with comprehensive knowledge of HIV/AIDS - Women': 'hiv_knowledge_women',
    'Prevalence of hypertension among men and women age 15-49 years  - Men': 'hypertension_men',
    'Prevalence of hypertension among men and women age 15-49 years  - Women': 'hypertension_women',
    'Percentage of men and women age 15-49 years with high Blood sugar level (>140 mg/dl) (%)    - Men': 'high_blood_sugar_men',
    'Percentage of men and women age 15-49 years with high Blood sugar level (>140 mg/dl) (%)    - Women': 'high_blood_sugar_women'
})
```

```python
#imr data cleaning
imr.head()
```

```python
imr.isnull().sum()
```

```python
#Flatten column names
imr.columns = ['_'.join([str(i) for i in col]).strip() for col in imr.columns]
#Reset index
imr = imr.reset_index()
#Check columns
imr.columns
```

```python
#Select required columns
imr_clean = imr[[ 'States/Union Territorries_Unnamed: 1_level_1', '1961_Male']].copy()
```

```python
#Rename columns
imr_clean.columns = ['state', 'imr']
```

```python
#Clean state names
imr_clean['state'] = imr_clean['state'].astype(str).str.strip().str.title()
```

```python
#Remove invalid rows
imr_clean = imr_clean.dropna(subset=['state'])
```

```python
#Reset index
imr_clean = imr_clean.reset_index(drop=True)
```

```python
#check
imr_clean.head()
```

```python
imr_clean = imr_clean.drop(0)
```

```python
imr_clean.head()
```

```python
#beds data cleaning
beds.head()
```

```python
#remove top garbage rows
beds = beds.iloc[2:]
```

```python
#Reset index
beds = beds.reset_index(drop=True)
beds.head()
```

```python
#Rename columns manually
beds_clean = beds.iloc[:, [1, 5, 9]].copy()
beds_clean.columns = ['state', 'hospitals', 'beds']
```

```python
#Clean state names
beds_clean['state'] = beds_clean['state'].astype(str).str.strip().str.title()
```

```python
#Convert to numbers
beds_clean['hospitals'] = pd.to_numeric(beds_clean['hospitals'], errors='coerce')
beds_clean['beds'] = pd.to_numeric(beds_clean['beds'], errors='coerce')
```

```python
#Remove invalid rows
beds_clean = beds_clean.dropna(subset=['state'])
beds_clean = beds_clean.reset_index(drop=True)
```

```python
beds_clean.head()
```

```python
#GSDP Data Cleaning
```

```python
#GSDP Data Cleaning
gsdp.head()
```

```python
#Select required columns
gsdp_clean = gsdp[['State Name', 'NSDP Per Capita (Nominal)2019-20']].copy()
```

```python
#Rename columns
gsdp_clean.columns = ['state', 'gsdp']
```

```python
#Remove ₹ and commas
gsdp_clean['gsdp'] = (gsdp_clean['gsdp'].astype(str).str.replace('₹', '').str.replace(',', ''))
```

```python
#Convert to numeric
gsdp_clean['gsdp'] = pd.to_numeric(gsdp_clean['gsdp'], errors='coerce')
```

```python
#Clean state names
gsdp_clean['state'] = gsdp_clean['state'].astype(str).str.strip().str.title()
```

```python
#Remove null rows
gsdp_clean = gsdp_clean.dropna(subset=['state', 'gsdp'])
gsdp_clean = gsdp_clean.reset_index(drop=True)
```

```python
#Check
gsdp_clean.head()
```

```python
#Coords Data Cleaning
coords.head()
```

```python
#required columns
coords_clean = coords[['state', 'latitude', 'longitude']].copy()
```

```python
#Standardize state names
coords_clean['state'] = coords_clean['state'].astype(str).str.strip().str.title()
```

```python
#Ensure numeric values
coords_clean['latitude'] = pd.to_numeric(coords_clean['latitude'], errors='coerce')
coords_clean['longitude'] = pd.to_numeric(coords_clean['longitude'], errors='coerce')
```

```python
#Check
coords_clean.head()
```

```python
print("All files Data cleaned Successfully")
```

```python
print("IMR:", imr_clean.shape)
print("Beds:", beds_clean.shape)
print("GSDP:", gsdp_clean.shape)
print("Coords:", coords_clean.shape)
```

```python
#Common state name fixes
state_fix = {
    'Andaman And Nicobar Islands': 'Andaman & Nicobar Islands',
    'Dadra And Nagar Haveli And Daman And Diu': 'Dadra & Nagar Haveli And Daman & Diu',
    'Nct Of Delhi': 'Delhi',
    'A & N Islands': 'Andaman & Nicobar Islands'
}

for df in [imr_clean, beds_clean, gsdp_clean, coords_clean]:
    df['state'] = df['state'].replace(state_fix)
```

```python
#Merge datasets
master = imr_clean.merge(beds_clean, on='state', how='left') \
                  .merge(gsdp_clean, on='state', how='left') \
                  .merge(coords_clean, on='state', how='left')
```

```python
master.head()
```

```python
master.shape
```

```python
#Check missing values
master.isnull().sum()
```

```python
#Fill missing numeric values
master['hospitals'] = master['hospitals'].fillna(0)
master['beds'] = master['beds'].fillna(0)
master['gsdp'] = master['gsdp'].fillna(master['gsdp'].median())
```

```python
master[master['latitude'].isnull()][['state']]
```

```python
#Remove invalid rows
master = master.dropna(subset=['state'])
master = master[master['state'] != 'All India']
```

```python
#Fix state name mismatch
state_fix = {
    'Jammu & Kashmir': 'Jammu And Kashmir',
    'Dadra & Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Daman & Diu': 'Dadra And Nagar Haveli And Daman And Diu',
    'Chhatisgarh': 'Chhattisgarh'
}

master['state'] = master['state'].replace(state_fix)
coords_clean['state'] = coords_clean['state'].replace(state_fix)
```

```python
#Merge coordinates again
master = master.drop(columns=['latitude', 'longitude'], errors='ignore')

master = master.merge(coords_clean[['state', 'latitude', 'longitude']],on='state',how='left')
```

```python
master['state'] = master['state'].astype(str).str.strip()

master = master[
    (master['state'].str.lower() != 'nan') &
    (master['state'] != '')
].reset_index(drop=True)
```
